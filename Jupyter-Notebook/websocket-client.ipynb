```python
#客户端连接至本地端口
from websocket import create_connection
import json

ws = create_connection("ws://localhost:12345/")
print("Sending 'init'...")
ws.send('{"id":0,"type":3}')
print("Sent")
print("Receiving...")
result =  ws.recv()
s = json.dumps(json.loads(result), indent=2)
print("Received '%s'" % s)
ws.close()
```
